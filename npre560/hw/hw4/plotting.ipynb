```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("plotting.mplstyle")
import cmath
from math import pi
```

```python
omegas_real = np.linspace(-5,5,1000)
omegas = np.array([complex(imag=x) for x in omegas_real])
```

```python
Lam = 1e-4     #s
beta = 0.0065
lam = 0.08     #s^-1

def Z(s):
    return 1/s * (Lam + beta / (s + lam))**-1

def amp(s):
    A = Z(s)
    Re = np.array([x.real for x in A])
    Im = np.array([x.imag for x in A])
    return np.sqrt(Re**2 + Im**2)

def phi(s):
    A = Z(s)
    Re = np.array([x.real for x in A])
    Im = np.array([x.imag for x in A])
    return np.arctan2(Im, Re)
```

```python
A = amp(omegas)
B = phi(omegas)
```

```python
fig, ax = plt.subplots()
ax.plot(omegas_real,A)
ax.set_xlabel(r'$\omega$')
ax.set_ylabel(r'$|Z(i\omega)|$')
fig.savefig('figs/magnitude.png')
```

```python
fig, ax = plt.subplots()
ax.plot(omegas_real,B)
ax.set_xlabel(r'$\omega$')
ax.set_ylabel(r'$\text{arg}[Z(i\omega)]$')
fig.savefig('figs/phase.png')
```
