```python
import matplotlib.pyplot as plt
import numpy as np
```

## Problem 1

```python
T_final = 0.2
n_steps = 1000
times = np.linspace(0,T_final, n_steps)
rho = 50 * np.ones(len(times)) * 1e-5

P0 = 1.0
Lambda = 2e-5
```

### a

```python
Pi = P0
Ps = []
dt = np.diff(times)[0]
for i in range(len(times)):
    Ps += [Pi]
    Pim1 = Pi
    A = 1 - (rho[i] * dt / Lambda) 
    Pi = Pim1 / A
```

```python
fig, ax = plt.subplots()
ax.plot(times, Ps)
ax.set_xlabel('Power')
ax.set_ylabel('Time [s]')
```

```python
fig.savefig('1-a.pdf')
```

### b

```python
Pi = P0
#beta = 750 * 1e-5
#beta = 100 * 1e-5
#beta = 50 * 1e-5
beta = 20 * 1e-5
C = P0 * beta / Lambda
Ps = []
dt = np.diff(times)[0]
for i in range(len(times)):
    Ps += [Pi]
    Pim1 = Pi
    A = 1 - ((rho[i] - beta) * dt / Lambda) 
    Pi = (Pim1 + C*dt)/ A
```

```python
fig, ax = plt.subplots()
ax.plot(times, Ps)
ax.set_xlabel('Power')
ax.set_ylabel('Time [s]')
```

```python
fig.savefig('1b4.pdf')
```
